```python
import pandas as pd

df = pd.read_csv("pstw_dataset.csv", sep=";")
```

```python
#df.info()
dataset = df
def null_avg(procent, column):
    return procent/100 * len(column)
for series_name, series in df.items():
    #print(series_name)l
    if series.isna().sum() > null_avg(50, series):
        #print(series_name,series.isna().sum(), round(series.isna().sum()/len(series)*100))
        dataset = dataset.drop(columns=[series_name])
        
dataset = pd.get_dummies(dataset, columns=["Improved Public Service"], dtype=int)
dataset = dataset.drop(columns=["PSTW ID","Name","Website","Description"])
print(dataset.info())
print(dataset["Geographical coverage (country)"])
```

```python
import numpy as np
import matplotlib.pyplot as plt

country_set = set(dataset["Geographical coverage (country)"])
country_count = []
name_count = {}
country_count.append(0)
for country in country_set:
    country_count.append( np.sum(dataset["Geographical coverage (country)"] == country))
    name_count[country] = np.sum(dataset["Geographical coverage (country)"] == country)
    print(name_count)
    
fig, ax = plt.subplots()
#ax.bar = (len(country_set), country_count, width=1, linewidth=0.7)
#ax.set(xlim=(0, len(country_set)), xticks=np.arange(0, len(country_set)),
#       ylim=(0, max(country_count)+1) , yticks=np.arange(0, max(country_count)+1))
#ax.set = (Xlabel="Countries", Ylabel="Count", title="Count of projects per country")
bplot = ax.bar(len(country_set), country_count)
ax.set_xlabel("Countries")
ax.set_ylabel("Count")  
ax.set_title("Count of projects per country")
#ax.set_xticks(np.arange(1, len(country_set)))
ax
plt.show()  
```

```python
fig, ax = plt.subplots()
print(type(name_count))
#print(type(name_count[0]))
#print(name_count[0][list(name_count[0].keys())[0]])
ax.bar(list(country_set), country_count[1:], width=0.4, linewidth=0.7)
ax.set_xlabel("Countries")
ax.set_ylabel("projects")
ax.set_title("Count of projects per country")
plt.xticks(rotation=90)
top_countries = sorted(name_count.items(), key=lambda item: item[1], reverse=True)[:10]



top_country_count = []
for country in top_countries:
    top_country_count.append(np.sum(dataset["Geographical coverage (country)"] == country))
print(top_countries)

top_countries_list = []
top_countries_count_list = []
for country in top_countries:
    top_countries_list.append(country[0])
    top_countries_count_list.append(country[1])
    

fig2, ax2 = plt.subplots()
ax2.bar(top_countries_list, top_countries_count_list)
ax2.set_xlabel("Countries")
ax2.set_ylabel("projects")
ax2.set_title("Top 10 countries with most projects")
plt.xticks(rotation=90)
```

```python
func_count = {}
for func in set(dataset["Functions of Government (COFOG level II)"]):
        func_count[func] = np.sum(dataset["Functions of Government (COFOG level II)"] == func)
gov_func = sorted(func_count.items(), key=lambda item: item[1], reverse=True)[:10]
top_func_list = []
top_func_count_list = []
for func in gov_func:
    top_func_list.append(func[0])
    top_func_count_list.append(func[1])
print(func_count)
print(gov_func)
fig2, ax2 = plt.subplots()
ax2.pie(func_count.values(), labels=list(func_count.keys()), autopct='%1.1f%%', startangle=90)
fig, ax = plt.subplots()
ax.pie(top_func_count_list, labels=top_func_list, autopct='%1.1f%%', startangle=90)
ax.set_title("Top 10 Government Functions")     
plt.show()


fig, ax3 = plt.subplots(1, 2)
ax3[0].pie(func_count.values(), labels=list(func_count.keys()), autopct='%1.1f%%', startangle=90)
ax3[0].set_title("All Government Functions")
ax3[1].pie(top_func_count_list, labels=top_func_list, autopct='%1.1f%%', startangle=90)
ax3[1].set_title("Top 10 Government Functions")
```
